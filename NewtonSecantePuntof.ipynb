```python
import math
import pylab as plt
nwrp=[]
oor=[] #ordenadas
eaproxn=[] #error aproximado newton
eaproxs=[] #error aproximado secante
erel=[] #error relativo newton
seca=[]
x=0
i=-2
oor=[] #oordenadas
f1=[]
f2=[]
eje=[]
e=[]
ee=[]
```

```python
def f(x):
    f=((math.e)**(-x))-x
    return f

def df(x):
    df=(-(math.e)**(-x))-1
    return df
    
def nr(x):
    nr=0
    nr=x-(f(x)/df(x))
    return nr

def arr():
    i=0
    xi=0
    while(i<11):
        xi=nr(xi)
        nwrp.append(xi)
        if(i>0):                                      
            oor.append(i-1)
            eap=ea(nwrp[i],nwrp[i-1])
            eaproxn.append(eap)
            er=et(xi)
            erel.append(er)
            print "i "," ","newton rapshon","\t ","e rel","\t \t ","e aprox"
            print repr(i).rjust(2), repr(xi).rjust(20),repr(erel[i-1]).rjust(20),repr(eaprox[i-1]).rjust(20)
        i+=1        
    
    
        
def ea(actual,ant):    
    ea=abs((actual-ant)/actual)*100
    return ea

def et(x):
    real=0.56714329    
    et=abs((real-nr(x))/real)*100
    return et
```

```python
def sec(x):
    sec=(x-( f(x)*(x-(x-1)) )/(f(x)-f(x-1)))
    return sec
    
def arr2():
    i=0
    s=1
    while(i<11):
        s=sec(s)
        seca.append(s)
        #oor.append(i)
        if(i>0):                                      
            #oor.append(i-1)
            eap=ea(seca[i],seca[i-1])
            eaproxs.append(eap)
            er=et(s)
            erel.append(er)
            print "i "," ","secante","\t ","e rel","\t \t ","e aprox"
            print repr(i).rjust(2), repr(s).rjust(20),repr(erel[i-1]).rjust(20),repr(eaprox[i-1]).rjust(20)
            #print "eaprox ",eaprox[i-1],"tamaño aprox", len(eaprox),"tamaño oor", len(oor)
        i=i+1
```
