```python
import math
import numpy as np
import pylab as pl
import matplotlib.patches as mpatches
h = 0.5#input()
x = -5
```

```python
def f(x,h):  
    f = (math.e**x) - 1
    return f

def range(x,h):
    r=x
    rango=[]
    while( r <= 5  ):        
        rango.append(r)
        r=r+h
    return rango
```

```python
def fanalitica(x,h):
    analitica=[]
    while(x<=5):        
        fanalitica = (math.e**(x))
        analitica.append(fanalitica)
        x=x+h
    return analitica

def fatras(x,h):
    atras=[]
    y=x
    while(y<=5):
        fatras = ( f(y,h) - f((y-h),h) ) / h
        atras.append(fatras)
        y=y+h
    return atras
        
def fadelante(x,h):
    adelante=[]
    y=x
    while(y<=5):        
        fadelante = (f(y+h,h)-f(y,h)) / ((y+h)-y)
        adelante.append(fadelante)
        y=y+h
    return adelante    

def fcentrada(x,h):
    centrada=[] 
    y=x
    while(y<=5):        
        fcentrada = (f(y+h,h)-f(y-h,h)) / (2*h)
        centrada.append(fcentrada)
        y=y+h
    return centrada  
    
def errVerL():
    evl=[]    
    i=0    
    canalitica = []    
    catras = []
    canalitica = fanalitica(x,h)    
    catras = fatras(x,h)  
    while(i<len(canalitica)):        
        errVL = canalitica[i]-catras[i]        
        evl.append(errVL)        
        i=i+1        
    return evl

def errVerN(): 
    evn=[]    
    i=0    
    canalitica = []    
    cadelante = []
    canalitica = fanalitica(x,h)    
    cadelante = fadelante(x,h)
    while(i<len(canalitica)):        
        errVN = canalitica[i]-cadelante[i]        
        evn.append(errVN)        
        i=i+1        
    return evn

def errVerC(): 
    evc=[]    
    i=0    
    canalitica = []    
    cadelante = []
    canalitica = fanalitica(x,h)    
    ccentrada = fcentrada(x,h)
    while(i<len(canalitica)):        
        errVC = canalitica[i]-ccentrada[i]        
        evc.append(errVC)        
        i=i+1        
    return evc

def errRL():
    erl=[] #error relativo atrÃ¡s
    cevl=[]
    cevl=errVerL()
    canalitica = fanalitica(x,h)
    i=0 
    while(i<len(canalitica)): 
        errRl=cevl[i]/canalitica[i]
        erl.append(errRl)  
        i=i+1   
    return erl

def errRN():
    erN=[] #error relativo adelante
    cevn=[]
    cevn=errVerN()
    canalitica = fanalitica(x,h)
    i=0 
    while(i<len(canalitica)): 
        errRn=cevn[i]/canalitica[i]
        ern.append(errRn)  
        i=i+1   
    return ern

def errRC():
    erc=[] #error relativo centrado
    cevc=[]
    cevc=errVerC()
    canalitica = fanalitica(x,h)
    i=0 
    while(i<len(canalitica)): 
        errRc=cevc[i]/canalitica[i]
        erc.append(errRc)  
        i=i+1   
    return erc
```

```python
greal = pl.plot(range(x,h), fanalitica(x,h), "r", label="Funcion analitica")
gatras = pl.plot(range(x,h), fatras(x,h), "b",label="Funcion atras")
gadelante = pl.plot(range(x,h), fadelante(x,h), "y",label="Funcion adelante" )
gcentrada = pl.plot(range(x,h), fcentrada(x,h), "k",label="Funcion centrada")
pl.title("Primera derivada")
pl.xlabel("Rango")
pl.ylabel("Funcion")
pl.xlim(-3.0, 8)
pl.ylim(-50.0, 200)
pl.legend()
pl.show()
```
