```python
import math
import numpy as np
import pylab as pl
```

```python
g=9.8
m=68.1
c=12.5

def tiempo():
    t=[]
    i=1
    while (i<=15):
        t.append(i)
        i=i+1
    return t

t=tiempo()

def modelo1():  
    mdl1=[]
    t=1
    while (t<=15):
        v1=(g*m/c)*(1-(math.e**((-c/m)*t)))
        mdl1.append(v1)
        t=t+1
    return mdl1

modelo1=modelo1()

def modelo2(): 
    mdl2=[]
    t=1
    while (t<=15):
        v2=(g*m/c)*(t/(3.75+t))
        mdl2.append(v2)
        t=t+1
    return mdl2

modelo2=modelo2()

def a1Mdl1():
    a1=0
    sumxy=0
    sumx=0
    sumy=0
    sumxx=0
    n=len(t)
    x=[]
    x=t
    y=[]
    y=modelo1
    i=0
    while (i<15):
        sumxy=(x[i]*y[i])+sumxy
        sumx=x[i]+sumx        
        sumy=y[i]+sumy            
        sumxx=(x[i]**2)+sumxx
        i=i+1
    a1=((n*sumxy)-(sumx*sumy))/((n*sumxx)-(sumx**2))
    return a1

a1Mdl1=a1Mdl1()

def a1Mdl2():
    a1=0
    sumxy=0
    sumx=0
    sumy=0
    sumxx=0
    n=len(t)
    x=[]
    x=t
    y=[]
    y=modelo2
    i=0
    while (i<15):
        sumxy=(x[i]*y[i])+sumxy
        sumx=x[i]+sumx        
        sumy=y[i]+sumy            
        sumxx=(x[i]**2)+sumxx
        i=i+1
    a1=((n*sumxy)-(sumx*sumy))/((n*sumxx)-(sumx**2))
    return a1

a1Mdl2=a1Mdl2()

def a0Mdl1():
    n=len(t)
    x=[]
    x=t
    y=[]
    y=modelo1
    promy=sum(y)/n
    promx=sum(x)/n
    a0=0
    a1=(a1Mdl1)
    a0=promy-(a1*promx)
    return a0

a0Mdl1=a0Mdl1()

def a0Mdl2():
    n=len(t)
    x=[]
    x=t
    y=[]
    y=modelo2
    promy=sum(y)/n
    promx=sum(x)/n
    a0=0
    a1=(a1Mdl2)
    a0=promy-(a1*promx)
    return a0
    
a0Mdl2=a0Mdl2()    
    
def ecLineal1():
    i=0
    x=t
    y=[]
    a_01=a0Mdl1
    a_11=a1Mdl1
    while (i<15):
        y1=a_01+x[i]*a_11
        y.append(y1)
        i=i+1
    return y

ecLineal1=ecLineal1()

def ecLineal2():
    i=0
    a_02=a0Mdl2    
    a_12=a1Mdl2
    x=t
    y=[]
    while (i<15):
        y2=a_02+x[i]*a_12
        y.append(y2)
        i=i+1
    return y

ecLineal2=ecLineal2()

print a1Mdl1
print a1Mdl2
print a0Mdl1
print a0Mdl2
print len(ecLineal1), len(ecLineal2)
```
